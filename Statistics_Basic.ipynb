# **Theoretical Questions**


1. What is statistics, and why is it important?
* Statistics is the discipline of collecting, analyzing, interpreting, and presenting data to discover patterns and inform decision-making, making it crucial for understanding variability, making predictions, and guiding policy and research across various fields.

2. What are the two main types of statistics?
* The two main types are descriptive statistics, which summarize and describe the features of a dataset, and inferential statistics, which use sample data to make generalizations and predictions about a larger population.

3. What are descriptive statistics?
* Descriptive statistics are techniques used to describe and summarize data through measures such as mean, median, mode, range, variance, and standard deviation, and through graphical representations like histograms and box plots.

4. What is inferential statistics?
* Inferential statistics involves drawing conclusions about a population based on a sample, using methods like hypothesis testing, confidence intervals, and regression analysis to make predictions and assess relationships.

5. What is sampling in statistics?
* Sampling is the process of selecting a subset of individuals or observations from a larger population to estimate characteristics of the whole population without measuring every member.

6. What are the different types of sampling methods?
* Common sampling methods include simple random sampling, systematic sampling, stratified sampling, cluster sampling, and convenience sampling, each with specific procedures for selecting participants.

7. What is the difference between random and non-random sampling?
* Random sampling ensures every member of the population has an equal chance of selection, reducing bias, whereas non-random sampling involves selecting individuals based on convenience or specific criteria, which may introduce bias.

8. Define and give examples of qualitative and quantitative data.
* Qualitative data are non-numeric and describe categories or qualities (e.g., colors, labels), while quantitative data are numeric and measure quantities (e.g., height, weight, temperature).

9. What are the different types of data in statistics?
* The four levels of measurement are nominal (labels, e.g., gender), ordinal (rank order, e.g., survey ratings), interval (numeric scales with equal intervals, e.g., temperature in Celsius), and ratio (interval scales with a true zero, e.g., weight).

10. Explain nominal, ordinal, interval, and ratio levels of measurement.
* Nominal data classify without order, ordinal data involve ordered categories, interval data have equal intervals without a true zero, and ratio data have equal intervals with a meaningful zero, allowing full range arithmetic operations.

11. What is the measure of central tendency?
* Measures of central tendency are statistics that describe the center of a dataset, including mean, median, and mode.

12. Define mean, median, and mode.
* Mean is the arithmetic average of values, median is the middle value when data are ordered, and mode is the most frequently occurring value in the dataset.

13. What is the significance of the measure of central tendency?
* Central tendency measures provide a single value representing the typical or central point of a data distribution, facilitating comparison and summarization.

14. What is variance, and how is it calculated?
* Variance is the average of the squared deviations from the mean, calculated by summing squared differences between each data point and the mean, then dividing by the number of observations (or nâ€“1 for sample variance).

15. What is standard deviation, and why is it important?
* Standard deviation is the square root of variance, representing the average distance of data points from the mean, and is important for understanding data dispersion in the original units.

16. Define and explain the term range in statistics.
* Range is the difference between the maximum and minimum values in a dataset, providing a simple measure of data spread.

17. What is the difference between variance and standard deviation?
* Variance measures dispersion in squared units, while standard deviation measures dispersion in the original units, making interpretation more intuitive.

18. What is skewness in a dataset?
* Skewness quantifies the asymmetry of a distribution, indicating whether data are skewed left (negative skew) or right (positive skew).

19. What does it mean if a dataset is positively or negatively skewed?
* Positive skew indicates a long tail on the right with most data on the left, while negative skew indicates a long tail on the left with most data on the right.

20. Define and explain kurtosis.
* Kurtosis measures the "tailedness" of a distribution, indicating whether data have heavy tails (leptokurtic), light tails (platykurtic), or normal tails (mesokurtic).

21. What is the purpose of covariance?
* Covariance measures how two variables change together, indicating the direction of their linear relationship, though its magnitude depends on the units of the variables.

22. What does correlation measure in statistics?
* Correlation quantifies the strength and direction of a linear relationship between two variables, standardized between -1 and 1 for comparability.

23. What is the difference between covariance and correlation?
* Covariance indicates the direction of the relationship in original units, while correlation standardizes this measure to a unitless value between -1 and 1, allowing comparison across different variable pairs.

24. What are some real-world applications of statistics?
* Statistics is applied in fields such as medicine (clinical trials), finance (risk modeling), social sciences (survey analysis), engineering (quality control), and sports (performance analytics) to inform decisions and predictions.

```python
# Q1: How do you calculate the mean, median, and mode of a dataset
import statistics as stats

data = [10, 20, 20, 30, 40, 40, 40, 50]
mean_val = stats.mean(data)
median_val = stats.median(data)
mode_val = stats.mode(data)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
# Q2:  Write a Python program to compute the variance and standard deviation of a dataset
import statistics as stats

data = [5, 7, 8, 9, 10]
variance = stats.pvariance(data)
stdev = stats.pstdev(data)
print("Variance:", variance)
print("Standard Deviation:", stdev)
```

```python
# Q4:  Implement sampling techniques like random sampling and stratified sampling
import pandas as pd

df = pd.DataFrame({
    'Group': ['A']*5 + ['B']*5,
    'Value': range(10)
})
simple = df.sample(n=4, random_state=0)
print("Simple random sample:\n", simple)

stratified = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(n=2, random_state=0))
print("\nStratified sample:\n", stratified)
```

```python
# 18: Implement Z-score normalization and explain its significance
''' Z-score normalization rescales your data so that it has a mean of 0 and a standard deviation of 1.
By converting raw values into standard deviations from the mean, it makes features with different units or scales directly comparable.
'''
import numpy as np

data = np.array([10, 20, 30, 40, 50], dtype=float)
mean_val = data.mean()
std_val = data.std(ddof=0)
z_scores = (data - mean_val) / std_val
print("Original data:", data)
print("Z-scores:", z_scores)
```

```python
# Q24: Implement the Pearson and Spearman correlation coefficients for a dataset
import numpy as np
from scipy.stats import pearsonr, spearmanr

x = np.array([10, 20, 30, 40, 50])
y = np.array([12, 24, 33, 47, 52])
pearson_corr, pearson_p = pearsonr(x, y)
print(f"Pearson correlation: {pearson_corr:.3f}, p-value: {pearson_p:.3f}")
spearman_corr, spearman_p = spearmanr(x, y)
print(f"Spearman correlation: {spearman_corr:.3f}, p-value: {spearman_p:.3f}")
```
